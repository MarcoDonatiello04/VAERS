```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. CARICAMENTO DEL DATASET CON TARGET
print("📂 Caricamento dataset con target...")
df = pd.read_csv('dataset/DATASET_WITH_TARGET.csv')

# 2. SEPARAZIONE FEATURE (X) E TARGET (y)
X = df.drop('SERIOUS', axis=1)
y = df['SERIOUS']

# 3. SUDDIVISIONE STRATIFICATA (80% Train, 20% Test)
# stratify=y garantisce che il rapporto del 20.44% sia mantenuto in entrambi i set
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

# 4. VERIFICA DELLE PROPORZIONI
print("\n" + "="*45)
print("⚖️ VERIFICA BILANCIAMENTO (Target: 1)")
print(f"Dataset Originale: {y.mean():.2%}")
print(f"Training Set:      {y_train.mean():.2%}")
print(f"Test Set:          {y_test.mean():.2%}")
print("="*45)

# 5. SALVATAGGIO DEI 4 FILE DEFINITIVI
print("\n💾 Salvataggio dei file separati...")
X_train.to_csv('dataset/X_train_final.csv', index=False)
X_test.to_csv('dataset/X_test_final.csv', index=False)
y_train.to_csv('dataset/y_train_final.csv', index=False)
y_test.to_csv('dataset/y_test_final.csv', index=False)

print(f"✅ Successo! Train set: {len(X_train)} righe | Test set: {len(X_test)} righe")
```
