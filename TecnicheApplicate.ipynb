```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import joblib

# 1. CARICAMENTO DEI FILE (Dopo la suddivisione stratificata)
print("📂 Caricamento dati per il raffinamento finale...")
X_train = pd.read_csv('dataset/X_train_final.csv')
X_test = pd.read_csv('dataset/X_test_final.csv')

# --- A. NUMDAYS: LOG TRANSFORMATION ---
# Usiamo log1p (cioè log(1+x)) per gestire i valori 0 (reazioni immediate)
# Il logaritmo di 0 è indefinito, log(1+0) è 0.
print("📉 Applicazione Log Transformation a NUMDAYS...")
X_train['NUMDAYS'] = np.log1p(X_train['NUMDAYS'])
X_test['NUMDAYS'] = np.log1p(X_test['NUMDAYS'])

# --- B. AGE_YRS: STANDARDIZZAZIONE (Z-SCORE) ---
# Media 0, Deviazione Standard 1
print("⚖️ Applicazione Standardizzazione (Z-Score) a AGE_YRS...")
scaler_age = StandardScaler()
X_train['AGE_YRS'] = scaler_age.fit_transform(X_train[['AGE_YRS']])
X_test['AGE_YRS'] = scaler_age.transform(X_test[['AGE_YRS']])

# --- C. NUM_SYMPTOMS: MIN-MAX SCALING [0,1] ---
# Scala i valori tra 0 e 1 mantenendo le proporzioni
print("📏 Applicazione Min-Max Scaling a NUM_SYMPTOMS...")
scaler_symptoms = MinMaxScaler()
X_train['NUM_SYMPTOMS'] = scaler_symptoms.fit_transform(X_train[['NUM_SYMPTOMS']])
X_test['NUM_SYMPTOMS'] = scaler_symptoms.transform(X_test[['NUM_SYMPTOMS']])

# 2. VERIFICA STATISTICA
print("\n" + "="*45)
print("📊 CONTROLLO TRASFORMAZIONI (Train Set)")
print("="*45)
statistiche = X_train[['NUMDAYS', 'AGE_YRS', 'NUM_SYMPTOMS']].describe().loc[['mean', 'std', 'min', 'max']]
print(statistiche)

# 3. SALVATAGGIO FILE TRASFORMATI
X_train.to_csv('dataset/X_train_transformed.csv', index=False)
X_test.to_csv('dataset/X_test_transformed.csv', index=False)

# Salviamo gli scaler per il futuro (es. per predire nuovi singoli pazienti)
joblib.dump(scaler_age, 'scaler_age_zscore.pkl')
joblib.dump(scaler_symptoms, 'scaler_symptoms_minmax.pkl')

print("\n✅ Trasformazioni completate e file salvati in 'dataset/'!")
```
