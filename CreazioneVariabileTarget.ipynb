```python
import pandas as pd
import numpy as np

# 1. CARICAMENTO DEL DATASET PULITO
print("📂 Caricamento dati da Originale2.csv...")
df = pd.read_csv('dataset/DATASET.csv')

# 2. DEFINIZIONE DELLE COLONNE DI GRAVITÀ (Standard FDA/VAERS)
# Queste colonne indicano se l'evento ha avuto un esito clinico rilevante
colonne_gravita = ['DIED', 'L_THREAT', 'HOSPITAL', 'DISABLE', 'ER_ED_VISIT']

# Controlliamo quali di queste colonne sono effettivamente presenti nel file
presenti = [col for col in colonne_gravita if col in df.columns]
print(f"🔍 Colonne cliniche rilevate: {presenti}")

# 3. CREAZIONE DELLA VARIABILE TARGET 'SERIOUS'
# Logica: Se almeno una colonna ha 'Y', il caso è GRAVE (1). Altrimenti è LIEVE (0).
# Usiamo .fillna('N') perché nel VAERS le celle vuote significano "No"
if presenti:
    df['SERIOUS'] = df[presenti].fillna('N').apply(
        lambda x: 1 if any(val == 'Y' for val in x) else 0, axis=1
    )

    # 4. STATISTICHE SUL TARGET
    totale = len(df)
    gravi = df['SERIOUS'].sum()
    lievi = totale - gravi
    percentuale_gravi = (gravi / totale) * 100

    print("\n" + "="*40)
    print(f"📊 DISTRIBUZIONE TARGET 'SERIOUS':")
    print(f"✅ Casi Lievi (0): {lievi}")
    print(f"🚨 Casi Gravi (1): {gravi}")
    print(f"📈 Rapporto: {percentuale_gravi:.2f}% di casi gravi")
    print("="*40)

    # 5. RIMOZIONE DELLE COLONNE SORGENTE (Prevenzione Data Leakage)
    # Fondamentale: il modello NON deve vedere queste colonne durante l'addestramento!
    df_finale = df.drop(columns=presenti)
    print(f"✂️ Colonne {presenti} rimosse per evitare Data Leakage.")

    # 6. SALVATAGGIO DEL DATASET CON TARGET
    output_path = 'dataset/DATASET_WITH_TARGET.csv'
    df_finale.to_csv(output_path, index=False)
    print(f"\n💾 Dataset pronto salvato in: {output_path}")

else:
    print("❌ Errore: Nessuna colonna di gravità trovata nel dataset!")
```
