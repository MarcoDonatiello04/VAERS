```python
import pandas as pd

# ======================================================
# 1Ô∏è‚É£ PATH
# ======================================================

TRAIN_PATH = "/data/interim/npl/train.csv"

# ======================================================
# 2Ô∏è‚É£ CARICAMENTO
# ======================================================

df = pd.read_csv(TRAIN_PATH)
print("Dataset caricato:", df.shape)

# ======================================================
# 3Ô∏è‚É£ FEATURE BINARIE
# ======================================================

binary_cols = [c for c in df.columns if df[c].dropna().isin([0,1]).all()]
print(f"\nFeature binarie totali: {len(binary_cols)}")

# ======================================================
# 4Ô∏è‚É£ STATISTICHE DI BASE
# ======================================================

stats = []

for col in binary_cols:
    ones_ratio = df[col].mean()
    stats.append((col, ones_ratio))

stats_df = pd.DataFrame(stats, columns=["feature", "ones_ratio"])
stats_df["zeros_ratio"] = 1 - stats_df["ones_ratio"]

# ======================================================
# 5Ô∏è‚É£ REPORT
# ======================================================

print("\nüìä Feature sempre ZERO:")
print(stats_df[stats_df["ones_ratio"] == 0]["feature"].tolist())

print("\nüìä Feature sempre UNO:")
print(stats_df[stats_df["ones_ratio"] == 1]["feature"].tolist())

print("\nüìä Feature molto RARE (<0.1%):")
print(stats_df[stats_df["ones_ratio"] < 0.001].sort_values("ones_ratio").head(10))

print("\nüìä Feature molto FREQUENTI (>99.9%):")
print(stats_df[stats_df["ones_ratio"] > 0.999].sort_values("ones_ratio", ascending=False).head(10))

print("\nüìä Distribuzione percentuale:")
print(stats_df["ones_ratio"].describe())
```
