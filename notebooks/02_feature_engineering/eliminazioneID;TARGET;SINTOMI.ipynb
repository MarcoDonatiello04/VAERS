```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# ======================================================
# 1Ô∏è‚É£ PATH
# ======================================================

BASE_PATH = "/Users/marcodonatiello/PycharmProjects/JupyterProject/data/"

TRAIN_PATH = BASE_PATH + "train.csv"
VAL_PATH   = BASE_PATH + "val.csv"
TEST_PATH  = BASE_PATH + "test.csv"

OUT_TRAIN = BASE_PATH + "train_step3_ready.csv"
OUT_VAL   = BASE_PATH + "val_step3_ready.csv"
OUT_TEST  = BASE_PATH + "test_step3_ready.csv"

# ======================================================
# 2Ô∏è‚É£ LOAD DATASET
# ======================================================

train_df = pd.read_csv(TRAIN_PATH)
val_df   = pd.read_csv(VAL_PATH)
test_df  = pd.read_csv(TEST_PATH)

print("Dataset caricati:")
print("Train:", train_df.shape)
print("Val  :", val_df.shape)
print("Test :", test_df.shape)

# ======================================================
# 3Ô∏è‚É£ STEP 3.0 ‚Äî DROP + X/y
# ======================================================

TARGET = "IS_SEVERE"
DROP_COLS = ["VAERS_ID", "LISTA_SINTOMI"]

def prepare_xy(df):
    df = df.drop(columns=DROP_COLS, errors="ignore")
    X = df.drop(columns=[TARGET])
    y = df[TARGET]
    return X, y

X_train, y_train = prepare_xy(train_df)
X_val, y_val     = prepare_xy(val_df)
X_test, y_test   = prepare_xy(test_df)

# ======================================================
# 4Ô∏è‚É£ STEP 3.1 ‚Äî SCALING SOLO NUMERICHE
# ======================================================

NUMERIC_COLS = ["AGE_YRS", "NUMDAYS", "NUMERO_SINTOMI"]

scaler = StandardScaler()
scaler.fit(X_train[NUMERIC_COLS])   # üëà SOLO TRAIN

X_train[NUMERIC_COLS] = scaler.transform(X_train[NUMERIC_COLS])
X_val[NUMERIC_COLS]   = scaler.transform(X_val[NUMERIC_COLS])
X_test[NUMERIC_COLS]  = scaler.transform(X_test[NUMERIC_COLS])

print("Scaling completato")

# ======================================================
# 5Ô∏è‚É£ RICOSTRUZIONE DATASET FINALI
# ======================================================

train_out = X_train.copy()
train_out[TARGET] = y_train

val_out = X_val.copy()
val_out[TARGET] = y_val

test_out = X_test.copy()
test_out[TARGET] = y_test

# ======================================================
# 6Ô∏è‚É£ SALVATAGGIO
# ======================================================

train_out.to_csv(OUT_TRAIN, index=False)
val_out.to_csv(OUT_VAL, index=False)
test_out.to_csv(OUT_TEST, index=False)

print("STEP 3 COMPLETATO ‚úÖ")
print("Salvati:")
print(OUT_TRAIN)
print(OUT_VAL)
print(OUT_TEST)
```
